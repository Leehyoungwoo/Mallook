```python
def extract_product_images(image_path):
    # 이미지 읽기
    image = cv2.imread(image_path)

    # 이미지 이진화
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY_INV)

    # 컨투어 검출
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 컨투어 필터링 및 상품 영역 추출
    product_images = []
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 1000:  # 예시적인 면적 필터링
            x, y, w, h = cv2.boundingRect(contour)
            product_images.append(image[y:y+h, x:x+w])

    return product_images


def extract_product_images_from_url(image_url, response):
    img_array = np.frombuffer(response.content, dtype=np.uint8)
    image = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

    # 이미지 이진화
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY_INV)

    # 컨투어 검출
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 컨투어 필터링 및 상품 영역 추출
    product_images = []
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 2000:  # 예시적인 면적 필터링
            x, y, w, h = cv2.boundingRect(contour)
            product_images.append(image[y:y+h, x:x+w])
    
    return product_images
```

```python
# 'hiver' 컬렉션에서 상품 5개의 'detail_images' 필드 가져오기
products = db.hiver.find({}, {'detail_images': 1}).limit(5)

# 결과 출력
for product in products:
    print('\n################################################################# 상품 구분선 ###########################################################\n')

    best_product_img = None
    product_images = []

    if 'detail_images' in product and product['detail_images']:
        for image_url in product['detail_images']:
            print(image_url)
            response = requests.get(image_url)
            product_images += extract_product_images_from_url(image_url, response)
            
    num_images = len(product_images)
    num_cols = 4  
    num_rows = max((num_images + num_cols - 1) // num_cols, 1)  
    
    plt.figure(figsize=(15, 3 * num_rows))
    
    # 모든 추출된 상품 이미지 표시
    for i, product_img in enumerate(product_images):
        plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(cv2.cvtColor(product_img, cv2.COLOR_BGR2RGB))
        plt.title(f'Result Image {i + 1}', fontsize=15)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

    # 가장 적합한 상품 이미지 선택 로직 (예시: 가장 큰 영역을 가진 이미지)
    # 여기서는 가장 큰 면적의 이미지를 가장 적합한 이미지로 선택합니다.
    max_area = 0
    for product_img in product_images:
        area = product_img.shape[0] * product_img.shape[1]
        if area > max_area:
            max_area = area
            best_product_img = product_img
    
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(best_product_img, cv2.COLOR_BGR2RGB))
    plt.title('Best Product Image', fontsize=15)
    plt.axis('off')
    plt.show()
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

train_images = train_images / 255.0

test_images = test_images / 255.0

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=10)

probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])

predictions = probability_model.predict(test_images)

def plot_image(i, predictions_array, true_label, img):
  true_label, img = true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  true_label = true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```
